#Thống kê các loại biển trong dataset

```python
def thong_ke(path):
    bien_do = 0
    bien_ngoai_giao = 0
    bien_trang = 0
    bien_vang = 0
    bien_xanh = 0
    for i in os.listdir(path):
        f = open(path + '/' + i, "r")
        content = f.read()
        lst = content.split('\n')
        for i in lst:
          temp = i.split()
          if len(temp) == 0:
            continue
          if temp[0] == '0':
            bien_do += 1
          elif temp[0] == '1':
            bien_ngoai_giao += 1
          elif temp[0] == '2':
            bien_trang += 1
          elif temp[0] == '3':
            bien_vang += 1
          else:
            bien_xanh += 1
          f.close()
    return bien_do, bien_ngoai_giao, bien_trang, bien_vang, bien_xanh 
```

* Thống kê các loại biển trong tập train

```python
bien_do_train, bien_ngoai_giao_train, bien_trang_train, bien_vang_train, bien_xanh_train = thong_ke('/content/labels/train') 
print('Số lượng nhãn trong tập train là:')
print(f'Biển đỏ: {bien_do_train}')
print(f'Biển ngoại giao: {bien_ngoai_giao_train}')
print(f'Biển trắng: {bien_trang_train}')
print(f'Biển vàng: {bien_vang_train}')
print(f'Biển xanh: {bien_xanh_train}')
```

* Thống kê các loại biển trong tập valid

```python
bien_do_val, bien_ngoai_giao_val, bien_trang_val, bien_vang_val, bien_xanh_val = thong_ke('/content/labels/valid') 
print('Số lượng nhãn trong tập valid là:')
print(f'Biển đỏ: {bien_do_val}')
print(f'Biển ngoại giao : {bien_ngoai_giao_val}')
print(f'Biển trắng: {bien_trang_val}')
print(f'Biển vàng: {bien_vang_val}')
print(f'Biển xanh: {bien_xanh_val}')
```

* Thống kê các loại biển trong tập test

```python
bien_do_test, bien_ngoai_giao_test, bien_trang_test, bien_vang_test, bien_xanh_test = thong_ke('/content/labels/test') 
print('Số lượng nhãn trong tập train train là:')
print(f'Biển đỏ: {bien_do_test}')
print(f'Biển ngoại giao : {bien_ngoai_giao_test}')
print(f'Biển trắng: {bien_trang_test}')
print(f'Biển vàng: {bien_vang_test}')
print(f'Biển xanh: {bien_xanh_test}')
```

```python
print(f'Biển đỏ trong toàn bộ: {bien_do_test+bien_do_train+bien_do_val}')
print(f'Biển ngoại giao trong toàn bộ: {bien_ngoai_giao_test+bien_ngoai_giao_train+bien_ngoai_giao_val}')
print(f'Biển trắng trong toàn bộ: {bien_trang_test+bien_trang_train+bien_trang_val}')
print(f'Biển vàng trong toàn bộ: {bien_vang_test+bien_vang_train+bien_vang_val}')
print(f'Biển xanh trong toàn bộ: {bien_xanh_test+bien_xanh_train+bien_xanh_val}')
```
